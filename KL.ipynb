```python
import numpy as np
import matplotlib.pyplot as plt

import sklearn, sklearn.metrics
```

```python
N = 1000000
```

## PHI

```python
def mk_phi():
    phi = np.random.normal(scale=np.pi/4, size=(2*N))
    phi = phi[phi < np.pi]
    phi = phi[phi > -np.pi]
    phi = phi[:N]
    return phi
```

```python
phi = mk_phi()

plt.plot(phi)
plt.axhline(y=np.pi/2, color='r', linestyle='-')
plt.axhline(y=-np.pi/2, color='r', linestyle='-')
```

```python
def pdf(x):
    p = plt.hist(x, range=(-np.pi, np.pi), bins=360, density=True)[0]
    return p
```

```python
plt.plot(pdf(phi))
```

```python
def KL(P,Q):
    epsilon = 0.00001
    P = P + epsilon
    Q = Q + epsilon

    divergence = np.sum(P*np.log(P/Q))
    return divergence
```

```python
phi0 = np.random.normal(size=(N))
phi0a = np.random.normal(size=(N))

phi1 = np.random.normal(loc=0, scale=0.2, size=(N))
phi2 = np.random.normal(loc=-1, scale=0.2, size=(N))
phi3 = np.random.normal(loc=1, scale=0.2, size=(N))
```

```python
pdf_phi = pdf(phi)
```

```python
pdf_phi0 = pdf(phi0)
pdf_phi0a = pdf(phi0a)
```

```python
pdf_phi1 = pdf(phi1)
```

```python
pdf_phi2 = pdf(phi2)
```

```python
pdf_phi3 = pdf(phi3)
```

```python
KL(pdf_phi0, pdf_phi0a)
```

```python
KL(pdf_phi, pdf_phi0)
```

```python
KL(pdf_phi0, pdf_phi)
```

```python
KL(pdf_phi1, pdf_phi0)
```
